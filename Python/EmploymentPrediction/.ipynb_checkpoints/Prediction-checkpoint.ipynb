```python
import numpy as np
import pandas as pd
import pymysql
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score,recall_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号的乱码问题
```

```python
# 连接mysql数据库
try:
    mydb = pymysql.connect(host='localhost', user='root', passwd='123456', port=3306, database='employmentprediction')
    # 创建游标对象
    cursor = mydb.cursor()
    print('连接成功!初始化成功!')
except:
    print('连接出错!')
    exit(0)
```

```python
sql = "select * from trainingdata;"
data = pd.read_sql(sql, mydb)
```

```python
data.set_index('id', inplace=True)
data = data[~data['y1'].isin(['自主创业', '部队'])]
# data.drop(columns='x13',inplace=True)
# data.drop(columns='y2',inplace=True)

X = data.loc[:,
             ['x1'
            ,'x2'
            ,'x3'
            ,'x4'
            ,'x5'
            ,'x6'
            ,'x7'
            ,'x8'
            ,'x9'
            ,'x10'
            ,'x11'
            ,'x12'
            # ,'x13'
            ]
            ]

y = data.loc[:,
             ['y1'
            ,'y2'
            ]
            ]
```

```python
y.replace(['医疗卫生单位', '城镇社区', '其他教学单位', '其他教学单位', '其他事业单位', '机关'], '事业单位',
             inplace=True)
y.replace('出国、出境深造', '升学', inplace=True)
y.replace(['其他暂不就业', '不就业拟升学'], '待就业', inplace=True)
y.replace(['国家基层项目', '地方基层项目'], '事业单位', inplace=True)
y.replace(['科研设计单位', '科研助理', '高等学校'], '高校教科研事业', inplace=True)
y.replace(['国有企业', '三资企业'], '国有、三资企业', inplace=True)


# y.replace(['其他企业','国有、三资企业及事业单位','自由职业','高校教科研事业','升学'],'就业',inplace=True)

y['y1'].value_counts()

# X,y
```

```python
y1_label_encoder = LabelEncoder()
y['y1_label_encoded'] = y1_label_encoder.fit_transform(y['y1'])

y_1hot_encoder = OneHotEncoder()
y_1hot = y_1hot_encoder.fit_transform(y['y1_label_encoded'].values.reshape(-1, 1))

# x13_label_encoder = LabelEncoder()
# X['x13_label_encoded'] = x13_label_encoder.fit_transform(X['x13'])

# x13_1hot_encoder = OneHotEncoder()
# x13_1hot = x13_1hot_encoder.fit_transform(X['x13_label_encoded'].values.reshape(-1, 1))

# X_encoded = pd.concat(
#     [X.reset_index(drop=True), pd.DataFrame(x13_1hot.toarray(), columns=x13_label_encoder.classes_),
#      pd.DataFrame(y_1hot.toarray(), columns=y1_label_encoder.classes_)], axis=1)

X_encoded = X

y_encoded = pd.concat([y.reset_index(drop=True),pd.DataFrame(y_1hot.toarray(),columns=y1_label_encoder.classes_)],axis=1)

# X = data_encoded.drop(
#     ['x13', 'x13_label_encoded', 'y1', 'y1_label_encoded', 'y2'] + list(x13_label_encoder.classes_) + list(
#         y1_label_encoder.classes_), axis=1)

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# X_encoded,y_encoded

# 标准化
# scaler = StandardScaler()
# scaled_X = scaler.fit_transform(X)
```

```python
# 创建并联数据
# ['其他企业', '升学', '国有、三资企业', '待就业', '自由职业', '高校教科研事业', '事业单位']
target_list=['其他企业', '升学', '国有、三资企业', '待就业', '自由职业', '高校教科研事业', '事业单位']
X_in = X_encoded.reset_index()
X_in.drop('id',axis=1,inplace=True)
print(X_in.dtypes)
# print(X_in.drop("id").var())
y_in = y_encoded[target_list]

X_y = pd.concat([X_in,y_in],axis=1)
X_y.dropna(inplace=True)
```

```python
# 主成分分析降维+居中
# pca = PCA(n_components=10)

# 创建列名列表
# x_pca_columns = {x:"x{}".format(x) for x in range(pca.n_components)}

# X6D = pca.fit_transform(X)
# X6D_df = pd.DataFrame(X6D).rename(columns=x_pca_columns)

X_in,y_in

# 随机抽样
# X_train, X_test, y_train, y_test = train_test_split(X_in, y_in, test_size=0.1, random_state=42)

# 分层抽样
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(X_y, y_in):
    strat_train_set = X_y.loc[train_index]
    strat_test_set = X_y.loc[test_index]

X_train=strat_train_set.drop((target_list), axis=1)
y_train=strat_train_set[target_list]
X_test=strat_test_set.drop((target_list), axis=1)
y_test=strat_test_set[target_list]

# 训练决策树模型
# model = KNeighborsClassifier()
# model = RandomForestClassifier(criterion='gini'
                               # ,splitter='random'
                              # )
model = KNeighborsClassifier(
                                # random_state=1
                               # ,criterion='log_loss'
                               # ,splitter='random'
                               # max_depth=12
                               # ,min_samples_leaf=1
                               # ,min_samples_split=3
                                
                              )
model.fit(X_train,y_train)

# y_train_pred = cross_val_predict(model, X_train, y_train, cv=10)
# y_train_score = cross_val_score(model, X_train, y_train, cv=10, scoring='accuracy')
# print("cross validation accuracy"+y_train_pred)
# confusion_mx = confusion_matrix(y_train[], y_train_pred)

# print(confusion_mx)

### 效果评估
# accuracy = accuracy_score(strat_test_set[list(y1_label_encoder.classes_)], y_pred)
# print("准确率:", accuracy)

# 预测测试集
# y_pred = model.predict(X_test)
# 预测训练集
# y_train_pre = model.predict(X_train)

# 计算测试集上的性能
test_score = model.score(X_test,y_test)
test_result = model.predict(X_test)

# 计算训练集上的性能
train_score = model.score(X_train,y_train)
train_result = model.predict(X_train)

print("训练集准确率",train_score)
print('测试集准确率',test_score)

test_matrix = confusion_matrix(test_result,y_test)
train_matrix = confusion_matrix(train_result,y_train)
precision=precision_score(test_result,y_test)
recall=recall_score(test_result,y_test)

print("测试集准确率","\n",precision)
print("测试机召回率","\n",recall)
print("测试集混淆矩阵","\n",test_matrix)
print("训练集混淆矩阵","\n",train_matrix)
```
